```python
count=0
for root, dirs, files in os.walk(os.path.join(BASE_PATH, 'train')):
    for file in files:
        if(count<1):
            count=count+1
            print(root,"\n", dirs,"\n", files,"\n", file)
        else:
            break
            
   
```

```python
files_path = [
    (root.replace("\\","/"), file) 
    for root, dirs, files in os.walk(os.path.join(BASE_PATH, 'train'))
    for file in files
]
```

```python
files_path[0]
```

```python
files_paths = list(map(lambda x: x.replace("\\","/"), files_path))
```

```python
files_paths[0].split('/')
```

```python
train_images_df = pd.DataFrame({
    'files_path': files_paths    
})
```

```python
train_images_df['family_id'] = train_images_df['files_path'].apply(lambda x: x.split('/')[3])
train_images_df['kin_id'] = train_images_df['files_path'].apply(lambda x: x.split('/')[4])
```

```python
train_images_df['excel_path'] = train_images_df[['family_id', 'kin_id']].apply(lambda x: '/'.join(x), axis=1)
```
